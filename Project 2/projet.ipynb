* **Moyenne**

```python
hrdataset['performancescore'].unique()
```

```python
score_mapping = {
    "Exceeds": 4,
    "Fully Meets": 3,
    "Needs Improvement": 2,
    "PIP": 1
}

hrdataset["performancescore_num"] = hrdataset["performancescore"].map(score_mapping)
```

```python
# Calcul des statistiques
stats = {
    "Moyenne": hrdataset['performancescore_num'].mean(),
    "Médiane": hrdataset['performancescore_num'].median(),
    "Écart-type": hrdataset['performancescore_num'].std(),
    "1er Quartile (Q1)": hrdataset['performancescore_num'].quantile(0.25),
    "2e Quartile (Q2)": hrdataset['performancescore_num'].quantile(0.50),
    "3e Quartile (Q3)": hrdataset['performancescore_num'].quantile(0.75)
}
```

Mettre sous forme de tableau

```python
stats_df = pd.DataFrame(stats, index=["Valeurs"])

stats_df
```

```python
sns.histplot(hrdataset['performancescore'], bins=10, kde=True)
plt.title("Distribution des scores de performance")
plt.show()
```

```python
sns.boxplot(x=hrdataset['performancescore_num'])
plt.title("Boxplot des scores de performance")
plt.xlabel("Score de performance (numérique)")
plt.show()
```

```python
numerical_columns = ["performancescore_num", "empsatisfaction", "salary", "absences"]

for col in numerical_columns:
    plt.figure(figsize=(8,5))
    sns.histplot(hrdataset[col], bins=10, kde=True)
    plt.title(f"Distribution de {col}")
    plt.xlabel(col)
    plt.ylabel("Nombre d'employés")
    plt.show()
```

```python
dept_counts = hrdataset["department"].value_counts()

plt.figure(figsize=(8,8))
plt.pie(dept_counts, labels=dept_counts.index, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
# Création d'un cercle blanc au centre pour effet "donut"
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Répartition des employés par département")
plt.tight_layout()
plt.show()
```
